```python
from sklearn import tree
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

file=open("woman.csv","rt",encoding="UTF8")

fileMatrix=[]
lineContent = file.readline()
while lineContent!='':
    fileMatrix.append(lineContent.strip('\n').split(','))
    lineContent = file.readline()
file.close()

#np.array([1,2,3]) 으로 ndarray 속성화 가능
#X는 att의 값들, y는 class att의 값들

X = [fileMatrix[i][1:3] for i in range(1,len(fileMatrix))]
y= [fileMatrix[i][3] for i in range(1,len(fileMatrix))]


for i in range(len(X)):
    X[i]=list(map(float,X[i]))


# 자동으로 데이터셋을 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

myTree = tree.DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=0)
myTree.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

y_pred_tr = myTree.predict(X_test)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred_tr))


from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
dot_data = export_graphviz(myTree, out_file=None, feature_names=['str', 'speed'],
                          class_names=["A","B","C"], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
